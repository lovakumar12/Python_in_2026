## what is a variable
 
 A variable is a named reference to a value stored in memory. It allows you to store data that you can use and manipulate throughout your program. Variables can hold different types of data, such as numbers, strings, lists, dictionaries, etc.

 Defining Variables
To define a variable in Python, you simply assign a value to a name using the = operator

```python
# Numeric variables
age = 25
height = 5.9

# String variables
name = "Alice"
greeting = "Hello, world!"

# Boolean variable
is_student = True

# List variable
fruits = ["apple", "banana", "cherry"]

# Dictionary variable
person = {"name": "Bob", "age": 30, "is_student": False}

# Tuple variable
coordinates = (10, 20)

# Set variable
unique_numbers = {1, 2, 3, 4, 5,"name"}
```

```python
## finding the memory of a variable
print(id(age))

## if we need  memory in octa and hexa format
print(hex(id(age)))
print(oct(id(age)))
```
